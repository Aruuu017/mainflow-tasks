```python
import matplotlib.pyplot as plt

# Step 1: Get value counts for the target column
target_counts = df['target'].value_counts()

# Step 2: Plot the bar chart with custom colors
target_counts.plot(kind='bar', color=["orchid", "salmon"])

# Step 3: Add titles and labels
plt.title('Distribution of Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Number of Cases')

# Step 4: Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Get value counts for the target column
target_counts = df['target'].value_counts()

# Step 2: Define labels and colors for the pie chart
labels = ['No Heart Disease', 'Heart Disease']
colors = ["orchid", "salmon"]

# Step 3: Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(target_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)

# Step 4: Add a title
plt.title('Distribution of Heart Disease (Pie Chart)')

# Step 5: Display the plot
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Calculate the number of males and females with heart disease
sex_counts = df[df['target'] == 1]['sex'].value_counts()
labels = ['Male', 'Female']

# Bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=labels, y=sex_counts.values, palette='pastel')
plt.title('Heart Disease Cases by Sex (Bar Chart)')
plt.ylabel('Number of Cases')
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(sex_counts, labels=labels, autopct='%1.1f%%', colors=['orchid', 'salmon'], startangle=140)
plt.title('Heart Disease Cases by Sex (Pie Chart)')
plt.axis('equal')
plt.show()
```

```python
# Calculate the number of people with each chest pain type
cp_counts = df['cp'].value_counts()

# Bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=cp_counts.index, y=cp_counts.values, palette='Blues_d')
plt.title('Distribution of Chest Pain Types (Bar Chart)')
plt.xlabel('Chest Pain Type')
plt.ylabel('Number of Cases')
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(cp_counts, labels=cp_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral', 'gold'])
plt.title('Distribution of Chest Pain Types (Pie Chart)')
plt.axis('equal')
plt.show()
```

```python
# Calculate the number of people with heart disease and each chest pain type
cp_heart_disease = df[df['target'] == 1]['cp'].value_counts()

# Bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=cp_heart_disease.index, y=cp_heart_disease.values, palette='OrRd')
plt.title('Heart Disease Cases by Chest Pain Type (Bar Chart)')
plt.xlabel('Chest Pain Type')
plt.ylabel('Number of Cases')
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(cp_heart_disease, labels=cp_heart_disease.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'darkorange', 'firebrick', 'salmon'])
plt.title('Heart Disease Cases by Chest Pain Type (Pie Chart)')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df is your DataFrame containing the heart disease data

# Question 7: Age Group Distribution in Heart Disease Cases
age_bins = [0, 30, 45, 60, 90]
age_labels = ['<30', '30-45', '46-60', '>60']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

age_group_heart_disease = df[df['target'] == 1]['age_group'].value_counts()

# Bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=age_group_heart_disease.index, y=age_group_heart_disease.values, palette='Blues')
plt.title('Heart Disease Cases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Cases')
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(age_group_heart_disease, labels=age_group_heart_disease.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'blue', 'darkblue', 'navy'])
plt.title('Heart Disease Cases by Age Group')
plt.axis('equal')
plt.show()


```

```python
# Question 8: Fasting Blood Sugar Levels
fbs_heart_disease = df[df['target'] == 1]['fbs'].value_counts()

# Bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=fbs_heart_disease.index, y=fbs_heart_disease.values, palette='Greens')
plt.title('Heart Disease Cases by Fasting Blood Sugar Level')
plt.xlabel('Fasting Blood Sugar (>120 mg/dl)')
plt.ylabel('Number of Cases')
plt.xticks([0, 1], ['<=120', '>120'])
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(fbs_heart_disease, labels=['<=120', '>120'], autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'darkgreen'])
plt.title('Heart Disease Cases by Fasting Blood Sugar Level')
plt.axis('equal')
plt.show()
```

```python
# Question 8: Resting Electrocardiographic Results
restecg_heart_disease = df[df['target'] == 1]['restecg'].value_counts()

# Bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=restecg_heart_disease.index, y=restecg_heart_disease.values, palette='Purples')
plt.title('Heart Disease Cases by Resting Electrocardiographic Results')
plt.xlabel('Resting ECG Result')
plt.ylabel('Number of Cases')
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(restecg_heart_disease, labels=restecg_heart_disease.index, autopct='%1.1f%%', startangle=140, colors=['plum', 'purple', 'violet'])
plt.title('Heart Disease Cases by Resting ECG Results')
plt.axis('equal')
plt.show()
```
